#### Dicionário de dados

* `tenure`:  meses de contrato do cliente
* `PhoneService`: assinatura de serviço telefônico
* `MultipleLines`: assisnatura de mais de uma linha de telefone
* `InternetService`: assinatura de um provedor internet
* `OnlineSecurity`: assinatura adicional de segurança online
* `OnlineBackup`: assinatura adicional de backup online
* `DeviceProtection`: assinatura adicional de proteção no dispositivo
* `TechSupport`: assinatura adicional de suporte técnico, menos tempo de espera
* `StreamingTV`: assinatura de TV a cabo
* `StreamingMovies`: assinatura de streaming de filmes
* `Contract`: tipo de contrato
* `PaperlessBilling`: se o cliente prefere receber online a fatura
* `PaymentMethod`: forma de pagamento
* `Charges.Monthly`: total de todos os serviços do cliente por mês
* `Charges.Total`: total gasto pelo cliente

Importando as bibliotecas necessárias

```python
import pandas as pd
import json
import requests
```

##<font color='green'>Entendendo quais informações o conjunto de dados possui</font>

Lendo os dados em formato json

```python
url = 'https://raw.githubusercontent.com/rodrigovitorino137/alura_challenges_ds1/main/Dados/Telco_Customer_Churn.json'
response = requests.get(url)
dados_brutos = response.json()
dados = pd.json_normalize(dados_brutos)
dados
```

Renomeando as colunas

```python
dados.columns
```

```python
dados = dados.rename(columns={'customerID': 'clienteID',
                                                  'customer.gender': 'genero',
                                                  'customer.SeniorCitizen': 'idoso',
                                                  'customer.Partner': 'parceiro',
                                                  'customer.Dependents': 'dependentes',
                                                  'customer.tenure': 'meses_contrato',
                                                  'phone.PhoneService': 'serviço_telefônico',
                                                  'phone.MultipleLines': 'multiplas_linhas',
                                                  'internet.InternetService': 'serviço_internet',
                                                  'internet.OnlineSecurity': 'segurança_online',
                                                  'internet.OnlineBackup': 'backup_online',
                                                  'internet.DeviceProtection': 'proteção_dispositivos',
                                                  'internet.TechSupport': 'suporte_tecnico',
                                                  'internet.StreamingTV': 'streaming_tv',
                                                  'internet.StreamingMovies': 'streaming_filmes',
                                                  'account.Contract': 'contrato',
                                                  'account.PaperlessBilling': 'faturamento_sem_papel',
                                                  'account.PaymentMethod': 'metodo_pagamento',
                                                  'account.Charges.Monthly': 'custos_mensais',
                                                  'account.Charges.Total': 'custos_totais'})
dados
```

```python
for col in dados.columns:
  dados[col] = dados[col].replace({'Yes':'Sim', 'No':'Nao', 'No phone service': 'Sem servico telefonico', 'No internet service': 'Sem servico internet'})

dados['genero'] = dados['genero'].replace({'Female':'Mulher', 'Male':'Homem'})

dados['serviço_internet'] = dados['serviço_internet'].replace('Fiber optic', 'Fibra optica')

dados['contrato'] = dados['contrato'].replace({'One year': 'Um ano', 'Month-to-month': 'Mes-a-mes', 'Two year': 'Dois anos'})

dados['metodo_pagamento'] = dados['metodo_pagamento'].replace({'Mailed check': 'Cheque enviado', 'Electronic check': 'Cheque eletronico', 'Credit card (automatic)': 'Cartao de credito(automatico)', 'Bank transfer (automatic)': 'Tranferencia bancaria(automatico)'})
```

```python
dados
```

Usando o método info() para saber o tipo dos dados em cada coluna e se existe a existencia de nan

```python
dados.info()
```

Concluímos que o tipo da coluna custos_totais está  errado, logo já sabemos que devemos converte-la para float.
Porém antes vamos verificar os dados únicos de cada coluna.

```python
for col in dados.columns:
  if dados[col].dtype == 'object':
    print(f"Coluna: {col}")
    print(dados[col].unique())
    print("-" * 30)
```

Agora sabemos que além de mudar o tipo da variavel da coluna custos_totais, devemos tratar strings vazias nessa coluna e na coluna Churn

##<font color='green'> Corrigindo as inconsistências nos dados<font color = 'green'>

Primeiro eliminamos as linhas onde a coluna churn está vazia.

```python
dados_sem_vazio = dados.query("Churn != ''").copy()
dados_sem_vazio.reset_index(drop = True, inplace = True)
```

Analizando com as strings vazias da coluna custos_totais

```python
index_erro = dados_sem_vazio['custos_totais'] == ' '
dados_sem_vazio[index_erro]
```

Verificamos que essa coluna não assume valor quando a coluna meses_contrato também é zero, ou seja, isso nos diz que o cliente não completou um mês de contrato, logo o mesmo só gastou no mês inicial. No lugar das strings vazias, iremos atribuir o valor da coluna custo_mensais.

```python
dados_sem_vazio.loc[index_erro, 'custos_totais'] = dados_sem_vazio.loc[index_erro, 'custos_mensais']
dados_sem_vazio.loc[index_erro]
```

Convertendo os dados da coluna custos_totais para float

```python
dados_sem_vazio['custos_totais'] = dados_sem_vazio['custos_totais'].astype(float)
```

Vereficando se existem dados duplicados

```python
dados_sem_vazio.duplicated().sum()
```

Confirmando a existencia de dados nulos

```python
dados_sem_vazio.isna().any().sum()
```

Inserindo uma coluna para nos dar os gastos diários, usando a coluna dos gastos mensais como base

```python
dados.insert(19, 'custo_diario', dados['custos_mensais']/30)
dados
```

Confirmando se está tudo ok.

```python
dados_sem_vazio.info()
```
